```python
import pandas as pd
b={'one':pd.Series([1,2,3],index=['b','a','c'])}
a=pd.DataFrame(b)
a['three']=pd.Series([5,6,7],index=['a','b','c'])
print(a)
a['four']=a['one']+a['three']
print(a)
del a['four']
print(a)
print(a.loc('a'))
```
